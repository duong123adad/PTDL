```python
import pandas as pd
import numpy as np
print(pd.__version__)
column_name  = ['Id','Name','Age','weight','m0006','m0612','m1218','f0006','f0612','f1218']
df = pd.read_csv('patient_heart_rate.csv',names=column_name)
df.head(10)
df[['FirstName','LastName']] = df['Name'].str.split(' ',expand=True)
df = df.drop('Name',axis=1)
df.head(5)

weight = df['weight']
for i in range(0, len(weight)):
    x = str(weight[i])
    if 'lbs' in x[-3]:
        x  = x[:-3:]
        float_x = float(x)
        y = int(float_x/2.2)
        weight[i] = y
        if 'kg' in x[-3]:
            x = x[:-3:]
            float_x = float(x)
            weight[i] = x
df.rename(columns={'weight':'Weight_kgs'},inplace=True)
df.head(5)
df.dropna(how='all', inplace=True)

df =df.drop_duplicates(subset=['FirstName','LastName','Age','Weight_kgs'], keep='first')

df.FirstName.replace({r'[^\x00-\x7F]+':''}, regex=True, inplace=True)
df.LastName.replace({r'[^\x00-\x7F]+':''}, regex=True, inplace=True)

print(df)

df['Age'] = df['Age'].fillna(df['Age'].mode()[0])

df['Weight_kgs'] = df['Weight_kgs'].astype(str).str.extract(r'(\d+\.?\d*)').astype(float)

weight_median = df['Weight_kgs'].median()
df['Weight_kgs'] = df['Weight_kgs'].fillna(weight_median)

print(df)


```

```python
if 'PulseRate' in df.columns:
    df = df.drop(columns=['PulseRate'])

df = pd.melt(df, 
             id_vars=['Id', 'FirstName', 'LastName', 'Age', 'Weight_kgs'], 
             value_name='PulseRate', 
             var_name='gender_time')

df_temp = df['gender_time'].str.extract(r"(\D)(\d+)(\d{2})", expand=True)
df_temp.columns = ['Gender', 'Lower_hour', 'Upper_hour']

df = pd.concat([df, df_temp], axis=1)
df = df.drop(['gender_time'], axis=1)

df = df.sort_values(['Id', 'Age', 'Weight_kgs', 'FirstName', 'LastName'])

print(df.head())
```

```python
df = df.replace('-',np.nan).dropna(subset=['PulseRate'])
df.head(10)
df['FirstName'].isnull().sum()
df['LastName'].isnull().sum()

df['FirstName'].fillna('Unknown', inplace=True)
df['LastName'].fillna('Unknown', inplace=True)

df= df.reset_index()
df.to_csv('patient_heart_rate_cleaned.csv', )
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('patient_heart_rate_cleaned.csv')
df.columns = df.columns.str.strip()

df['PulseRate'] = pd.to_numeric(df['PulseRate'], errors='coerce')
df = df.dropna(subset=['PulseRate'])

df['TimeGroup'] = df['Lower_hour'].astype(str).str.zfill(2) + '-' + df['Upper_hour'].astype(str).str.zfill(2)

pivot_df = df.pivot_table(values='PulseRate', index='TimeGroup', columns='Gender', aggfunc='mean')

pivot_df.plot(kind='bar', figsize=(12, 6), color=['pink', 'skyblue'], edgecolor='black')

plt.xlabel('Nhm gi')
plt.ylabel('Nhp tim trung bnh (BPM)')
plt.title('So snh nhp tim trung bnh theo nhm gi v gii tnh')
plt.legend(title='Gii tnh', labels=['N (f)', 'Nam (m)'])
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('patient_heart_rate_cleaned.csv')
df.columns = df.columns.str.strip()

df['PulseRate'] = pd.to_numeric(df['PulseRate'], errors='coerce')
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df = df.dropna(subset=['PulseRate', 'Age'])

df['Fullname'] = df['FirstName'].fillna('') + ' ' + df['LastName'].fillna('')

patient_stats = df.groupby('Fullname').agg({
    'Age': 'mean',
    'PulseRate': 'mean'
}).reset_index()

plt.figure(figsize=(10, 6))
plt.scatter(patient_stats['Age'], patient_stats['PulseRate'], color='darkblue', s=100, alpha=0.7, edgecolors='white')

for i, txt in enumerate(patient_stats['Fullname']):
    plt.annotate(txt, (patient_stats['Age'][i], patient_stats['PulseRate'][i]), xytext=(5, 5), textcoords='offset points', fontsize=9)

plt.xlabel(' tui trung bnh')
plt.ylabel('Nhp tim trung bnh (BPM)')
plt.title('Tng quan gia  tui v Nhp tim trung bnh ca tng bnh nhn')
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('patient_heart_rate_cleaned.csv')
df.columns = df.columns.str.strip()

df['PulseRate'] = pd.to_numeric(df['PulseRate'], errors='coerce')
df = df.dropna(subset=['PulseRate'])

df['TimeGroup'] = df['Lower_hour'].astype(str).str.zfill(2) + '-' + df['Upper_hour'].astype(str).str.zfill(2)

time_gender_stats = df.groupby(['TimeGroup', 'Gender'])['PulseRate'].mean().unstack()

plt.figure(figsize=(12, 6))
plt.plot(time_gender_stats.index, time_gender_stats['m'], marker='o', label='Nam (Male)', color='skyblue', linewidth=2)
plt.plot(time_gender_stats.index, time_gender_stats['f'], marker='s', label='N (Female)', color='pink', linewidth=2)

plt.xlabel('Khung gi')
plt.ylabel('Nhp tim trung bnh (BPM)')
plt.title('S thay i nhp tim trung bnh theo khung gi v gii tnh')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.6)

plt.tight_layout()
plt.show()
```
